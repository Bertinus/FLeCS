```python
environment = ["all env.", "observational", "Cebpa KO", "Cebpe KO", "both KO"]

column_1_float = [100., 20.0, 18.0, 21.5, 0.26  , 0.53, 0.013, 0.003, 0.007]
# [100., 19.9, 18.0, 21.4, 0.25, 0.5, 0.01, 3.5, 4.0]

my_cmap = plt.get_cmap("Pastel1")
width = 0.3  # the width of the bars
fig, ax = plt.subplots(layout='constrained', figsize=(7, 2.5))

rects = ax.bar(0., column_1_float[0], width, edgecolor = "black", color=my_cmap(0.))
ax.bar_label(rects, padding=-15)

rects = ax.bar(1., column_1_float[1], width, label=environment[1], edgecolor = "black", color=my_cmap(0.2))
ax.bar_label(rects, padding=3)

rects = ax.bar([1.5, 3., 5.], [column_1_float[2], column_1_float[4], column_1_float[7]], width, label=environment[2], edgecolor = "black", color=my_cmap(0.4))
ax.bar_label(rects, padding=3)

rects = ax.bar([2., 3.5, 5.5], [column_1_float[3], column_1_float[5], column_1_float[8]], width, label=environment[3], edgecolor = "black", color=my_cmap(0.6))
ax.bar_label(rects, padding=3)

rects = ax.bar(4., column_1_float[6], width, label=environment[4], edgecolor = "black", color=my_cmap(1.))
ax.bar_label(rects, padding=3)

ax.set_xticks([0, 1.5, 3.5, 5.25], ["ATACseq network\nedges", "Non-zero edges\n(pval < 1/#edges)", "Affected edges\n(pval < 1/#edges)", "Change of sign\ndue to KO"])

plt.ylabel("percentage among \nnetwork edges")
ax.legend(loc='upper right', ncols=2)
# plt.xticks(rotation=90)
plt.yscale("log")
plt.show()
```
